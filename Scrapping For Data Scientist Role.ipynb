**The Website We will be Scrapping is [Times Jobs] (https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Scientist&txtLocation=)**

```python
from bs4 import BeautifulSoup
```

```python
with open('data_scientist.html', 'r') as joy:
    page = joy.read()
```

```python
soup = BeautifulSoup(page, 'lxml')
```

```python
articles = soup.select('ul') #<-- select individual tags with css selector. Selecting each article tags in the html
articles
```

```python
result = soup.select('h3')  # select each h3 tags inside article tag  (this h3 tag contains the companies names in the html)
result
```

```python
company_names = [chioma.text.strip('\n \n \n (More Jobs)') for chioma in result]
company_names
```

```python
result = soup.select('span[class="srp-skills"]')
key_skills = [chioma.text.strip('\n \n \n') for chioma in result]
key_skills
```

```python
data_scientist = list(zip(company_names, key_skills))
list(zip(company_names, key_skills))
```
